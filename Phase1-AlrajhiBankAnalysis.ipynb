```python
import tweepy as tw
import pandas as pd
import csv 
```

```python
#Credential values to access Twitter API 
%store -r CONSUMER_KEY
%store -r CONSUMER_SECRET
%store -r ACCESS_TOKEN
%store -r ACCESS_SECRET
```

```python
def twitter_setup():
    auth = tw.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
    auth.set_access_token(ACCESS_TOKEN, ACCESS_SECRET)
    api = tw.API(auth, wait_on_rate_limit=True)
    return api
```

```python
extractor = twitter_setup()
result = []
```

```python
cursor = tw.Cursor(extractor.search_tweets, q='ุจูู ุงูุฑุงุฌุญู -filter:retweets', tweet_mode='extended').items(10000)
i = 0

for tweet in cursor:
    print(tweet.full_text)
    print('-----------------------------')
    i +=1
    print(i)
    result.append({'Tweet':tweet.full_text, 'Retweets': tweet.retweet_count,'Likes':tweet.favorite_count, 'Length': len(tweet.full_text) , 'Created At': tweet.created_at})
```
